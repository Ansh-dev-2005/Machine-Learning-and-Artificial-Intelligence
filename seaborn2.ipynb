```python

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Ansh-dev-2005/Datasets/main/World%20Happiness%20Report.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
sns.relplot(x="Country", y="Economy", data=df)
```

```python
sns.relplot(x="Country", y="Economy",hue="Happiness Score", data=df)
```

```python
sns.relplot(x="Country", y="Happiness Score", data=df)
```

```python
sns.relplot(x="Region", y="Happiness Score", data=df)
```

```python
sns.relplot(x="Region", y="Economy", data=df)
```

```python
sns.relplot(x="Country", y="Economy",hue="Region", data=df)
```

```python
sns.catplot(x="Region", y="Economy", data=df)
```

```python
sns.catplot(x="Country", y="Region", data=df)
```

```python
sns.catplot(x="Economy", y="Region", data=df)
```
