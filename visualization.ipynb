```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([1,2])
ypoints = np.array([5,10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1,20])
ypoints = np.array([10,200])
plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1,6])
ypoints = np.array([4,5])
plt.plot(xpoints, ypoints, 'X')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])
plt.plot(ypoints, '*')
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
plt.plot(xpoints, ypoints, marker = 'o', ms = 20, mec = 'b')
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
plt.plot(xpoints, ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'y')
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])
plt.plot(xpoints, ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'y', mew = 5)
plt.show()
```
