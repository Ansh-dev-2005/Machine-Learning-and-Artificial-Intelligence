```python
#### Outliers
df.boxplot(column=["lead_time"])
```

```python
def remove_outliers(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
low_leadtime,up_leadtime=remove_outliers(df["lead_time"])
df["lead_time"]=np.where(df["lead_time"]>up_leadtime,up_leadtime,df["lead_time"])
df["lead_time"]=np.where(df["lead_time"]<low_leadtime,low_leadtime,df["lead_time"])
df.boxplot(column=["lead_time"])
```
